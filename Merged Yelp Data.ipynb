```python
df_b = pd.read_json('Yelp Data/yelp_academic_dataset_business.json', lines=True)
```

```python
df_b.head()
```

```python
#Keeping businesses that are still open in the dataset
df_b = df_b[df_b['is_open']==1]
```

```python
#Dropping irrelevent Columns
drop_columns = ['hours','is_open','review_count']
df_b = df_b.drop(drop_columns, axis=1)
```

```python
df_b.head()
```

```python
#Checking how many categories of business & seperating them
df_explode = df_b.assign(categories = df_b.categories.str.split(', ')).explode('categories')
```

```python
df_explode.categories.value_counts()
```

```python
df_explode[df_explode.categories.str.contains('Restaurants',
                      case=True,na=False)].categories.value_counts()
```

```python
bus_Res = df_b[df_b['categories'].str.contains('Restaurants',case=False, na=False)]
bus_Tor_Res= bus_Res[bus_Res['city'].str.contains('Toronto',case=False, na=False)]
bus_Tor_Res.head()
```

```python
#df_r=pd.read_json('Yelp Data/yelp_academic_dataset_review.json', lines=True)
#chunks = pd.read_json('Yelp Data/yelp_academic_dataset_review.json', lines=True, chunksize = 100000)
#Identifying the data type of each column can reduce memory usage.
#Luckily, Yelp provided documentation for their dataset so we can set the data type of each column.
#Here, the chunk size of 1,000,000 means Pandas will be reading 1,000,000 lines each time. In this dataset of review.json, there are more than 6 million reviews (rows). Reducing the chunk size might be easier to load and check the results faster.
df_r='Yelp Data/yelp_academic_dataset_review.json'
size=100000
review=pd.read_json(df_r, lines=True,
                      dtype={'review_id':str,'user_id':str,
                              'business_id':str,'stars':int,
                             'date':str,'text':str,'useful':int,
                             'funny':int,'cool':int}, chunksize=size)
# There are multiple chunks to be read
chunk_list = []
for chunk_review in review:
    # Drop columns that aren't needed
    chunk_review = chunk_review.drop(['review_id','useful','funny','cool'], axis=1)
    # Renaming column name to avoid conflict with business overall star rating
    chunk_review = chunk_review.rename(columns={'stars': 'review_stars'})
    # Inner merge with edited business file so only reviews related to the business remain
    chunk_merged = pd.merge(bus_Tor_Res, chunk_review, on='business_id', how='inner')
    # Show feedback on progress
    print(f"{chunk_merged.shape[0]} out of {size:,} related reviews")
    chunk_list.append(chunk_merged)
# After trimming down the review file, concatenate all relevant data back to one dataframe
df = pd.concat(chunk_list, ignore_index=True, join='outer', axis=0)
```

```python
csv_name = "yelp_rev_TorRes.csv"
df.to_csv(csv_name, index=False)
```
